```python
import tensorflow as tf
import tensorflow_addons as tfa



# Define the optimizer
optimizer = tf.keras.optimizers.legacy.Adam(learning_rate=0.008)

num_epochs = 3
temperature = 0.5  # Temperature parameter for NT-Xent loss
cumulative_loss = 0.0

best_loss = float('inf')  # Initialize best loss to a very high value

# Assuming simclr_data_generator is an instance of SimCLRDataGenerator
for epoch in range(num_epochs):
    
    print(f"Epoch {epoch+1}/{num_epochs}")
# Generate the Training Batch
        # Instantiate the data generator
    simclr_train_generator = SimCLRDataGenerator(original_train_generator)
        # Generate augmented data
    augmented_data = simclr_train_generator.generate()

    batch_generator = simclr_encoder.process_batch(augmented_data)
    
    cumulative_loss = 0.0
    batch_index = 0

    while True:
        try:
            # Fetch the next batch of augmented images
            with tf.GradientTape() as tape:
                batch_index += 1
                projected_representations_1, projected_representations_2 = next(batch_generator)
                
                # Concatenate the representations along the batch dimension
                concatenated_representations = tf.concat([projected_representations_1, projected_representations_2], axis=0)

                # Compute the NT-Xent loss
                loss = nt_xent_loss(concatenated_representations, temperature)
                #loss = NT_Xent_tf(projected_representations_1, projected_representations_2, 0.5)

                # Compute gradients with respect to the trainable variables of the encoder and projection head
                gradients = tape.gradient(loss, simclr_encoder.encoder_model.trainable_variables + simclr_encoder.projection_head.trainable_variables)
                
            # Apply gradients to update the encoder and projection head
            optimizer.apply_gradients(zip(gradients, simclr_encoder.encoder_model.trainable_variables + simclr_encoder.projection_head.trainable_variables))
            
            cumulative_loss += loss.numpy()
            print(f"\tBatch {batch_index}, Loss: {loss.numpy()}, Average Loss : {cumulative_loss / batch_index}")
            
            # Save the model every 500 batches if cumulative_loss is smaller than the best_loss so far
            if (batch_index % 500) == 0 and (cumulative_loss < best_loss):
                best_loss = cumulative_loss
                simclr_encoder.save_models('models/encoder_model', 'models/projection_head_model')
                print(f"Saved models at batch {batch_index} with improved average loss: {cumulative_loss / batch_index}")


        except StopIteration:
            # Break the loop when the generator is exhausted
            break
```

## Function to reshape the validation dataset

```python
import os

base_dir = 'tiny-imagenet-200'
val_dir = os.path.join(base_dir, 'val')
annotations_file = os.path.join(val_dir, 'val_annotations.txt')
images_dir = os.path.join(val_dir, 'images')

val_annotations = {}
with open(annotations_file, 'r') as file:
    for line in file:
        parts = line.strip().split('\t')
        image_name, class_id = parts[0], parts[1]
        val_annotations[image_name] = class_id

for image, class_id in val_annotations.items():
    class_dir = os.path.join(val_dir, class_id)
    os.makedirs(class_dir, exist_ok=True)
    src = os.path.join(images_dir, image)
    dst = os.path.join(class_dir, image)
    shutil.move(src, dst)

shutil.rmtree(images_dir)
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.utils import to_categorical
import numpy as np
import os
import shutil

train_dir = os.path.join(dataset_path, 'train')
val_dir = os.path.join(dataset_path, 'val')

train_datagen = ImageDataGenerator(rescale=1./255)
val_datagen = ImageDataGenerator(rescale=1./255)
```

```python
# having issue with extra folder after renaming the folder names
train_folders = set(os.listdir(train_dir))
val_folders = set(os.listdir(val_dir))

non_matching_folders = val_folders - train_folders

if non_matching_folders:
    print("Non-matching folders found:", non_matching_folders)
    for folder in non_matching_folders:
        folder_path = os.path.join(val_dir, folder)
        shutil.rmtree(folder_path)
        print(f"Removed folder: {folder_path}")
else:
    print("Non-matching folders not found.")
```

```python
train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(64, 64),  
    batch_size=32,
    shuffle=True,
    class_mode='categorical')

val_generator = val_datagen.flow_from_directory(
    val_dir,
    target_size=(64, 64),
    batch_size=32,
    shuffle=True,
    class_mode='categorical')
```

```python
# as the contrastive learning model does not associate output with label
# the following function aims to make the connection

from tqdm import tqdm

def extract_features(model, generator):
    features = []
    labels = []
    # add progress tracker
    steps_per_epoch = generator.n // generator.batch_size

    for imgs, lbls in tqdm(generator, total=steps_per_epoch):
        batch_features = model.encoder_model.predict(imgs)
        features.append(batch_features)
        labels.append(lbls)
        if len(features) * generator.batch_size >= generator.samples:
            break

    return np.vstack(features), np.vstack(labels)
```

```python
# the above function requires this command
tf.data.experimental.enable_debug_mode()
```

```python
#extract features
train_features, train_labels = extract_features(simclr_encoder, train_generator)
val_features, val_labels = extract_features(simclr_encoder, val_generator)
```

```python
#modeling
num_classes = 200
input_shape = train_features.shape[1:]
classifier_input = Input(shape=input_shape)
x = Dense(512, activation='relu')(classifier_input)
x = Dense(256, activation='relu')(x)
output = Dense(num_classes, activation='softmax')(x)  
classifier_model = Model(inputs=classifier_input, outputs=output)

#final model
classifier_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
classifier_model.fit(train_features, train_labels, validation_data=(val_features, val_labels), epochs=5)
```
