```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# Creating a DataFrame for the provided data
data = {
    'Model': ['ResNet-152', 'ResNet-152', 'ResNet-152', 'ResNet-152', 'ResNet-152', 'ResNet-101', 'ResNet-50'],
    'Batch Size': [32, 64, 128, 256, 512, 512, 512],
    'Test Accuracy': [10.27, 9.9, 10, 10.44, 14.65, 21.7, 32]
}

df = pd.DataFrame(data)

# Plotting
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Batch Size', y='Test Accuracy', hue='Model', style='Model', s=100)
plt.title('Test Accuracy Comparison Across Models and Batch Sizes')
plt.xlabel('Batch Size')
plt.ylabel('Test Accuracy (%)')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Data for the plot
data = {
    'Model': ['ResNet-152', 'ResNet-152', 'ResNet-152', 'ResNet-152', 'ResNet-152', 'ResNet-101', 'ResNet-50','ResNet-101'],
    'Batch Size': [32, 64, 128, 256, 512, 512, 512,32],
    'Test Accuracy': [10.27, 9.9, 10, 10.44, 14.65, 21.7, 32,19.9]
}
df = pd.DataFrame(data)

# Creating the scatter plot with discrete x-axis values
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Batch Size', y='Test Accuracy', hue='Model', style='Model', s=100)

# Adjusting x-axis to be discrete
plt.xticks(df['Batch Size'].unique())

plt.title('Test Accuracy Comparison Across Models and Batch Sizes')
plt.xlabel('Batch Size')
plt.ylabel('Test Accuracy (%)')
plt.grid(True)
plt.show()
```
